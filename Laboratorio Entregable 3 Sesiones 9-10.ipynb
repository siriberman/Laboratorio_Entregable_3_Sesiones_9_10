```python
#Parte2: Limpieza de Datos con Python
#Corrección de errores
#•Eliminar o reemplazar valores nulos (fillna, dropna).
#•Eliminar duplicados con .drop_duplicates().
#•Normalización de formatos en columnas (e.j., convertir texto a MAYÚSCULAS, estandarizar fechas al formato AAAA/MM/DD con pd.to_datetime).

import pandas as pd
df = pd.read_csv('CSV Laboratorio 3 AD-B.csv')
df['City'] = df['City'].str.upper()
df['Name'] = df['Name'].str.upper()
df['Category'] = df['Category'].str.upper()
df['Category'] = df['Category'].str.replace('CAT', 'CATEGORY')
df['Category'] = df['Category'].str.replace('CATEGORYEGORY', 'CATEGORY')
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', format='%Y-%m-%d')
print("\nDatos después de la limpieza:")
print(df)
print("\nVerificación final:")
print(df.isnull().sum())
print(df.dtypes)
#exportar dataset a.csv
df.to_csv('CSV Laboratorio 3 AD-B.csv', index=False)
```
